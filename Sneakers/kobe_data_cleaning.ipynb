```python
# convert release date to date time string format
def release_date(s):
    if math.isnan(s): 
        return ""
    else: 
        return str(datetime.strptime(str(int(s)),"%Y%m%d").date())
```

```python
def color_comparison(color1, color2):
    if type(color1) == str and type(color2) == str:
        if len(color2) > len(color1):
            return color2
        if len(color1) >= len(color2):
            return color1
    else:
        if isNaN(color1) and isNaN(color2):
            return None
        if isNaN(color1) and not isNaN(color2):
            return color2
        if not isNaN(color1) and isNaN(color2):
            return color1
    
```
